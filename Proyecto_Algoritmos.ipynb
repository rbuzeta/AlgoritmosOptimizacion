```python
# En esta secci√≥n s√≥lo hay constantes del algoritmo, que no se deben contabilizar como parte de la exploraci√≥n porque se pueden poner en forma expl√≠cita.
# Es decir, se pueden reemplazar por 9 + 36 + 14 + 152 elementos constantes.
C = {1,2,3,4,5,6,7,8,9}

# ùí´  (productos sin repetici√≥n, y < x)
P = [(x,y) for x in C for y in C if y < x]          # |P| = 36

# ùíü  (divisiones exactas entre d√≠gitos distintos)
D = [(a,b) for a in C for b in C if a != b and a % b == 0]   # |D| = 14

# ‚Ñ≥  (x*y/z, todos distintos y z | x*y)   ‚Äî 152 elementos
M = [(x,y,z) for x in C for y in C for z in C
     if len({x,y,z}) == 3 and (x*y) % z == 0]

print(P)
print(len(P))  # 36

print(D)
print(len(D))  # 14

print(M)
print(len(M))  # 152
```

```python
# Para cada tupla guardaremos tambi√©n el valor num√©rico que produce, para evitar recalcular cada vez en el algoritmo:
P_val = {(x,y): x*y              for (x,y)     in P}
D_val = {(a,b):  a//b            for (a,b)     in D}
M_val = {(x,y,z): (x*y)//z       for (x,y,z)   in M}

print(P_val)
print(len(P_val))  # 36
print(D_val)
print(len(D_val))  # 14
print(M_val)
print(len(M_val))  # 152
```

```python
# Funci√≥n principal. Esta es la funci√≥n que se ejecuta por cada expresi√≥n buscada.
# Recibe un n√∫mero n y devuelve una expresi√≥n que produzca ese n√∫mero, o None si no existe tal expresi√≥n.
def busca_expresion(n: int) -> str | None:
    # ‚ñ∏ Estructura (1)  p + c ‚àí d
    for (x,y) in P:
        for c in C - {x,y}:
            for (a,b) in D:
                if len({x,y,c,a,b}) != 5:        # d√≠gitos no repetidos
                    continue
                if P_val[(x,y)] + c - D_val[(a,b)] == n:
                    return f"{x}*{y} + {c} - {a}/{b}"

    # ‚ñ∏ Estructura (2)  p + d ‚àí c
    for (x,y) in P:
        for (a,b) in D:
            if len({x,y,a,b}) < 4:
                continue
            for c in C - {x,y,a,b}:
                if P_val[(x,y)] + D_val[(a,b)] - c == n:
                    return f"{x}*{y} + {a}/{b} - {c}"

    # ‚ñ∏ Estructura (3)  c‚ÇÅ + c‚ÇÇ ‚àí m
    for c1 in C:
        for c2 in C - {c1}:
            for (x,y,z) in M:
                if len({c1,c2,x,y,z}) != 5:
                    continue
                if c1 + c2 - M_val[(x,y,z)] == n:
                    return f"{c1} + {c2} - {x}*{y}/{z}"

    # ‚ñ∏ Estructura (4)  c‚ÇÅ + m ‚àí c‚ÇÇ
    for c1 in C:
        for (x,y,z) in M:
            if c1 in {x,y,z}:
                continue
            for c2 in C - {c1,x,y,z}:
                if c1 + M_val[(x,y,z)] - c2 == n:
                    return f"{c1} + {x}*{y}/{z} - {c2}"

    # ‚ñ∏ Si llegamos aqu√≠, no existe t√©rmino que produzca n
    return None

print(busca_expresion(77))  # Ejemplo de uso
```

```python
# Vamos a estresar el algoritmo con todos los n√∫meros del -50 al 50
# y a imprimir el resultado de cada uno.
for n in range(-50, 51):
    expr = busca_expresion(n)
    if expr:
        print(f"{n:3d}:  {expr}")
    else:
        print(f"{n:3d}:  ‚Äî sin soluci√≥n ‚Äî")
```

```python
import time, statistics

# Se asume que busca_expresion(n:int) ‚Üí str|None
# ya est√° definida en el entorno actual.

def benchmark_busca_expresion(lo: int = 1, hi: int = 100) -> None:
    # Recorre los enteros [lo, hi], llama a busca_expresion(n) y muestra tiempos individuales m√°s un resumen estad√≠stico.

    tiempos   : list[float] = []
    encontrados: int        = 0

    print(f"{'n':>4} ‚îÇ {'Expresi√≥n can√≥nica':<35} ‚îÇ Tiempo (ms)")
    print("‚îÄ" * 60)

    for n in range(lo, hi + 1):
        t0 = time.perf_counter()
        expr = busca_expresion(n)          # ‚Üê funci√≥n ya existente
        dt  = time.perf_counter() - t0

        tiempos.append(dt)
        if expr is not None:
            encontrados += 1

        print(f"{n:4d} ‚îÇ { (expr or '‚Äî sin soluci√≥n ‚Äî')[:35]:<35} ‚îÇ {dt*1000:9.3f}")

    # Estad√≠sticas globales
    total = sum(tiempos)
    print("\n‚îÄ‚îÄ‚îÄ Estad√≠sticas ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ")
    print(f" Total consultados : {len(tiempos)}")
    print(f" Encontrados       : {encontrados}")
    print(f" Tiempo total      : {total*1000:9.2f} ms")
    print(f" Tiempo promedio   : {statistics.mean(tiempos)*1000:9.3f} ms")
    print(f" Desv. est√°ndar    : {statistics.stdev(tiempos)*1000:9.3f} ms")
    print(f" Tiempo m√≠nimo     : {min(tiempos)*1000:9.3f} ms")
    print(f" Tiempo m√°ximo     : {max(tiempos)*1000:9.3f} ms")


# ‚îÄ‚îÄ Ejemplo de ejecuci√≥n ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
benchmark_busca_expresion(1, 78)
```

```python
# -------------------------------------------------
#  eval_expr SIMPLE
# -------------------------------------------------
def eval_expr(digits: tuple[int, ...], ops: tuple[str, ...]):
# Construye la expresi√≥n d0 op0 d1 op1 d2 op2 d3 op3 d4, la eval√∫a con eval() y devuelve el entero resultante (si el valor es un entero exacto), None en cualquier otro caso

    expr = ''.join(f"{digits[i]}{ops[i]}" for i in range(4)) + str(digits[4])
    try:
        val = eval(expr)              # usa precedencia normal de Python
    except ZeroDivisionError:
        return None

    # Comprobamos que sea entero exacto (float o int)
    if isinstance(val, float):
        return int(val) if val.is_integer() else None
    return val                        # ya es int


# -------------------------------------------------
#  fuerza_bruta con el eval simplificado
# -------------------------------------------------
from itertools import combinations, permutations

OPS = ['+', '-', '*', '/']

def fuerza_bruta(n: int) -> str | None:
    """Devuelve la primera expresi√≥n can√≥nica que vale n, o None."""
    for digits_set in combinations(range(1, 10), 5):        # elijo 5 d√≠gitos distintos
        for d in permutations(digits_set):                  # orden de los d√≠gitos
            for ops in permutations(OPS):                   # orden de los 4 operadores
                val = eval_expr(d, ops)
                if val is not None and val == n:
                    return ''.join(f"{d[i]}{ops[i]}" for i in range(4)) + str(d[4])
    return None


print(f"Fuerza bruta para n=17: {fuerza_bruta(17)}")  # Ejemplo de uso
```

```python
def benchmark_fuerza_bruta(lo=1, hi=100):
    tiempos = []
    print(f"{'n':>4} ‚îÇ {'Expresi√≥n encontrada':<35} ‚îÇ Tiempo (ms)")
    print("‚îÄ"*60)
    for n in range(lo, hi+1):
        t0 = time.perf_counter()
        expr = fuerza_bruta(n)
        dt  = time.perf_counter() - t0
        tiempos.append(dt)
        print(f"{n:4d} ‚îÇ {(expr or '‚Äî sin soluci√≥n ‚Äî')[:35]:<35} ‚îÇ {dt*1000:9.3f}")
    print("\n‚îÄ‚îÄ Estad√≠sticas sobre {:d} valores ‚îÄ‚îÄ".format(hi-lo+1))
    print(f"  total   : {sum(tiempos)*1000:9.2f} ms")
    print(f"  promedio: {statistics.mean(tiempos)*1000:9.3f} ms")
    print(f"  m√≠nimo  : {min(tiempos)*1000:9.3f} ms")
    print(f"  m√°ximo  : {max(tiempos)*1000:9.3f} ms")

# ‚îÄ‚îÄ Llamada de prueba (1‚Ä¶78) ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
benchmark_fuerza_bruta(1, 78)
```
