```python
# En esta secciÃ³n sÃ³lo hay constantes del algoritmo, que no se deben contabilizar como parte de la exploraciÃ³n porque se pueden poner en forma explÃ­cita.
# Es decir, se pueden reemplazar por 9 + 36 + 14 + 152 elementos constantes.
C = {1,2,3,4,5,6,7,8,9}

# ğ’«  (productos sin repeticiÃ³n, y < x)
P = [(x,y) for x in C for y in C if y < x]          # |P| = 36

# ğ’Ÿ  (divisiones exactas entre dÃ­gitos distintos)
D = [(a,b) for a in C for b in C if a != b and a % b == 0]   # |D| = 14

# â„³  (x*y/z, todos distintos y z | x*y)   â€” 152 elementos
M = [(x,y,z) for x in C for y in C for z in C
     if len({x,y,z}) == 3 and (x*y) % z == 0]

print(P)
print(len(P))  # 36

print(D)
print(len(D))  # 14

print(M)
print(len(M))  # 152
```

```python
# Para cada tupla guardaremos tambiÃ©n el valor numÃ©rico que produce, para evitar recalcular cada vez en el algoritmo:
P_val = {(x,y): x*y              for (x,y)     in P}
D_val = {(a,b):  a//b            for (a,b)     in D}
M_val = {(x,y,z): (x*y)//z       for (x,y,z)   in M}

print(P_val)
print(len(P_val))  # 36
print(D_val)
print(len(D_val))  # 14
print(M_val)
print(len(M_val))  # 152
```

```python
# ------------------------------------------------------------------
#  Busca una expresiÃ³n vÃ¡lida (usa exactamente los 5 dÃ­gitos y los 4
#  operadores una sola vez) que produzca el entero n.  Si no existe,
#  devuelve None.
# ------------------------------------------------------------------
def busca_expresion(n: int) -> str | None:
    # â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
    # Estructura (1)  p + c - d
    # â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
    for (x, y) in P:                                   # p = x*y
        p_val = P_val[(x, y)]
        digits_p = {x, y}
        for c in C - digits_p:                         # dÃ­gito suelto
            digits_pc = digits_p | {c}
            for (a, b) in D:                           # d = a/b
                if len(digits_pc | {a, b}) != 5:
                    continue
                if p_val + c - D_val[(a, b)] == n:
                    return f"{x}*{y} + {c} - {a}/{b}"

    # â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
    # Estructura (2)  p + d - c
    # â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
    for (x, y) in P:
        p_val = P_val[(x, y)]
        digits_p = {x, y}
        for (a, b) in D:
            digits_pd = digits_p | {a, b}
            if len(digits_pd) < 4:
                continue
            for c in C - digits_pd:
                if p_val + D_val[(a, b)] - c == n:
                    return f"{x}*{y} + {a}/{b} - {c}"

    # â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
    # Estructura (3)  c + d - p
    # â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
    for c in C:
        for (a, b) in D:
            digits_cd = {c, a, b}
            if len(digits_cd) < 3:
                continue
            d_val = D_val[(a, b)]
            for (x, y) in P:
                if len(digits_cd | {x, y}) != 5:
                    continue
                if c + d_val - P_val[(x, y)] == n:
                    return f"{c} + {a}/{b} - {x}*{y}"

    # â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
    # Estructura (4)  c + p - d
    # â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
    for c in C:
        for (x, y) in P:
            digits_cp = {c, x, y}
            if len(digits_cp) < 3:
                continue
            p_val = P_val[(x, y)]
            for (a, b) in D:
                if len(digits_cp | {a, b}) != 5:
                    continue
                if c + p_val - D_val[(a, b)] == n:
                    return f"{c} + {x}*{y} - {a}/{b}"

    # â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
    # Estructura (5)  câ‚ + câ‚‚ - m
    # â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
    for c1 in C:
        for c2 in C - {c1}:
            digits_c = {c1, c2}
            for (x, y, z) in M:                        # m = x*y/z
                if len(digits_c | {x, y, z}) != 5:
                    continue
                if c1 + c2 - M_val[(x, y, z)] == n:
                    return f"{c1} + {c2} - {x}*{y}/{z}"

    # â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
    # Estructura (6)  câ‚ + m - câ‚‚
    # â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
    for c1 in C:
        for (x, y, z) in M:
            digits_cm = {c1, x, y, z}
            if len(digits_cm) < 4:
                continue
            m_val = M_val[(x, y, z)]
            for c2 in C - digits_cm:
                if c1 + m_val - c2 == n:
                    return f"{c1} + {x}*{y}/{z} - {c2}"

    # â€”â€”â€” No existe expresiÃ³n que produzca n â€”â€”â€”
    return None


print(busca_expresion(-69))  # Ejemplo de uso
```

```python
# Vamos a estresar el algoritmo con todos los nÃºmeros del -69 al 77
# y a imprimir el resultado de cada uno, demostrando que todos los nÃºmeros en ese rango
# tienen una expresiÃ³n vÃ¡lida.

salida = []

for n in range(-69, 78):
    expr = busca_expresion(n)
    if expr:
        salida.append(f"{n:3d}:  {expr}")
    else:
        salida.append(f"{n:3d}:  â€” sin soluciÃ³n â€”")

# Imprimir todo de una vez
print("  |  ".join(salida))
```

```python
import time, statistics

# Se asume que busca_expresion(n:int) â†’ str|None
# ya estÃ¡ definida en el entorno actual.

def benchmark_busca_expresion(lo: int = 1, hi: int = 100) -> None:
    # Recorre los enteros [lo, hi], llama a busca_expresion(n) y muestra tiempos individuales mÃ¡s un resumen estadÃ­stico.

    tiempos   : list[float] = []
    encontrados: int        = 0

    print(f"{'n':>4} â”‚ {'ExpresiÃ³n canÃ³nica':<35} â”‚ Tiempo (ms)")
    print("â”€" * 60)

    for n in range(lo, hi + 1):
        t0 = time.perf_counter()
        expr = busca_expresion(n)          # â† funciÃ³n ya existente
        dt  = time.perf_counter() - t0

        tiempos.append(dt)
        if expr is not None:
            encontrados += 1

        print(f"{n:4d} â”‚ { (expr or 'â€” sin soluciÃ³n â€”')[:35]:<35} â”‚ {dt*1000:9.3f}")

    # EstadÃ­sticas globales
    total = sum(tiempos)
    print("\nâ”€â”€â”€ EstadÃ­sticas â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€")
    print(f" Total consultados : {len(tiempos)}")
    print(f" Encontrados       : {encontrados}")
    print(f" Tiempo total      : {total*1000:9.2f} ms")
    print(f" Tiempo promedio   : {statistics.mean(tiempos)*1000:9.3f} ms")
    print(f" Desv. estÃ¡ndar    : {statistics.stdev(tiempos)*1000:9.3f} ms")
    print(f" Tiempo mÃ­nimo     : {min(tiempos)*1000:9.3f} ms")
    print(f" Tiempo mÃ¡ximo     : {max(tiempos)*1000:9.3f} ms")


# â”€â”€ Ejemplo de ejecuciÃ³n â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
benchmark_busca_expresion(1, 77)
```

```python
# -------------------------------------------------
#  eval_expr SIMPLE
# -------------------------------------------------
def eval_expr(digits: tuple[int, ...], ops: tuple[str, ...]):
# Construye la expresiÃ³n d0 op0 d1 op1 d2 op2 d3 op3 d4, la evalÃºa con eval() y devuelve el entero resultante (si el valor es un entero exacto), None en cualquier otro caso

    expr = ''.join(f"{digits[i]}{ops[i]}" for i in range(4)) + str(digits[4])
    try:
        val = eval(expr)              # usa precedencia normal de Python
    except ZeroDivisionError:
        return None

    # Comprobamos que sea entero exacto (float o int)
    if isinstance(val, float):
        return int(val) if val.is_integer() else None
    return val                        # ya es int


# -------------------------------------------------
#  fuerza_bruta con el eval simplificado
# -------------------------------------------------
from itertools import combinations, permutations

OPS = ['+', '-', '*', '/']

def fuerza_bruta(n: int) -> str | None:
    """Devuelve la primera expresiÃ³n canÃ³nica que vale n, o None."""
    for digits_set in combinations(range(1, 10), 5):        # elijo 5 dÃ­gitos distintos
        for d in permutations(digits_set):                  # orden de los dÃ­gitos
            for ops in permutations(OPS):                   # orden de los 4 operadores
                val = eval_expr(d, ops)
                if val is not None and val == n:
                    return ''.join(f"{d[i]}{ops[i]}" for i in range(4)) + str(d[4])
    return None


print(f"Fuerza bruta para n=17: {fuerza_bruta(17)}")  # Ejemplo de uso
```

```python
def benchmark_fuerza_bruta(lo=1, hi=100):
    tiempos = []
    print(f"{'n':>4} â”‚ {'ExpresiÃ³n encontrada':<35} â”‚ Tiempo (ms)")
    print("â”€"*60)
    for n in range(lo, hi+1):
        t0 = time.perf_counter()
        expr = fuerza_bruta(n)
        dt  = time.perf_counter() - t0
        tiempos.append(dt)
        print(f"{n:4d} â”‚ {(expr or 'â€” sin soluciÃ³n â€”')[:35]:<35} â”‚ {dt*1000:9.3f}")
    print("\nâ”€â”€ EstadÃ­sticas sobre {:d} valores â”€â”€".format(hi-lo+1))
    print(f"  total   : {sum(tiempos)*1000:9.2f} ms")
    print(f"  promedio: {statistics.mean(tiempos)*1000:9.3f} ms")
    print(f"  mÃ­nimo  : {min(tiempos)*1000:9.3f} ms")
    print(f"  mÃ¡ximo  : {max(tiempos)*1000:9.3f} ms")

# â”€â”€ Llamada de prueba (1â€¦78) â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
benchmark_fuerza_bruta(1, 77)
```
